```python
import pandas as pd
import config2 as cfg
from sqlalchemy import create_engine
```

```python
# creating a connection to an aws rds postgres cloud database.
rds_connection_string = f'postgres:{cfg.password}@test-db.cy2enoewwvsi.us-east-2.rds.amazonaws.com:5432/stocks_db'
engine = create_engine(f'postgres://{rds_connection_string}')
```

```python
# checking out the table names in the database.
engine.table_names()
```

```python
# checking out the SandP_500 table.
sandp = pd.read_sql_query('SELECT * FROM sandp', con=engine)
sandp.head()
```

```python
# checking out the sandp_russell data.
sandp_russell = pd.read_sql_query('SELECT * FROM sandp_russell', con=engine)
# dropping wierd column that showed up and dont want to do other process.
# sandp_russell = sandp_russell.drop(columns={'Unnamed: 0'})
sandp_russell.head()
```

```python
# checking out the russell_2000 data.
russell = pd.read_sql_query('SELECT * FROM russell_2000', con=engine)
russell.head()
```

```python
# checking out the new_route data.
new_route = pd.read_sql_query('SELECT * FROM new_route', con=engine)
new_route.head()
```

```python
# sandp returned about 146 from 2009/2010 to 2018/2019.
```

```python
dataframes = [sandp, sandp_russell, russell, new_route]
for df in dataframes:
    above_below_sandp = []
    for returns in df['percent_return_on_investment']:
        if returns > 228:
            above_below_sandp.append(1)

        else:
            above_below_sandp.append(0)

    df['above_below_sandp_return'] = above_below_sandp 
```
