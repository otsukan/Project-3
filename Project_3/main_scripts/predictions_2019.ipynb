```python
# importing necessary modules.
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
import config9 as cfg
import requests
```

```python
postgres_password = cfg.password
database_name = 'optimal_portfolio'
```

```python
# creating a connection to an aws rds postgres cloud database.
rds_connection_string = f'postgres:{postgres_password}@localhost:5432/{database_name}'
engine = create_engine(f'postgres://{rds_connection_string}')
```

```python
# checking out the table names in the database.
engine.table_names()
```

```python
# checking out the SandP_500 table.
sandp = pd.read_sql_query('SELECT * FROM sandp', con=engine)
sandp.head()
```

```python
symbols = []
for symbol in sandp['symbol']:
    symbols.append(symbol)
```

```python
response = requests.get(f'https://financialmodelingprep.com/api/v3/financials/income-statement/AAPL').json()
float(response['financials'][-2]['Revenue'])
```

```python
revenue = []
net_income = []
for ticker in symbols:
    try:
        
        response = requests.get(f'https://financialmodelingprep.com/api/v3/financials/income-statement/{ticker}').json()

        if response['financials'][-2]['date'].split('-')[0] == '2009' or response['financials'][-2]['date'].split('-')[0] == '2010':
            revenue.append(float(response['financials'][-2]['Revenue']))
            net_income.append(float(response['financials'][-2]['Net Income']))
            
            print(response['financials'][-2]['Revenue'])
            
        else:
            revenue.append(0)
            net_income.append(0)
            
    except:
        pass
```

```python
len(revenue)
```

```python
sandp['Revenue'] = revenue
sandp['Net Income'] = net_income
```

```python
sandp
```

```python
sandp = sandp[['industry', 'net_income_growth','short_term_coverage_ratio','eps_diluted_growth', 'gross_profit_margin', 'price_to_book', 'price_to_sales', 'price_to_free_cash_flow', 'asset_growth', 'Revenue', 'Net Income']]
```

```python
sandp = sandp.groupby('industry').mean()
sandp
```

```python
features = ['net_income_growth','short_term_coverage_ratio','eps_diluted_growth', 'gross_profit_margin', 'price_to_book', 'price_to_sales', 'price_to_free_cash_flow', 'asset_growth']
dataframe_names = []
for (column, header) in zip(sandp.columns, features):
    top = sandp.sort_values(by=column, ascending=False).head(33).reset_index()
    feature_names = []
    feature_value = []
    for (name, value) in zip(top['industry'], top[column]):
        feature_names.append(name)
        feature_value.append(value)

    df = pd.DataFrame(feature_value, feature_names).reset_index()
    df = df.rename(columns={0: header})
    dataframe_names.append(df)
```
